# Iris dataset Munging and Analysis

```python
# Importing the necessary libraries and setting Seaborn global values
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)
```

## Importing and Describing the Iris dataset
* Import the Iris dataset into a Pandas dataframe
* Describe the Iris Dataset
* Check if numeric features are missing.

```python
iris_df = pd.read_csv('Iris.csv')
iris_df.describe()
```

* We notice no missing values for the numeric features.
* Checking if the data is evenly distributed across the classes.

```python
iris_df["Species"].value_counts()
```

### Visualization of the Iris dataset
* Scatter plot of the Iris sppecies based on Sepal Characteristics

```python
sns.FacetGrid(iris_df, hue="Species", size=5).map(plt.scatter, "SepalLengthCm", "SepalWidthCm").add_legend()
plt.show()
```

* Scatter plot of the Iris sppecies based on Petal Characteristics

```python
sns.FacetGrid(iris_df, hue="Species", size=5).map(plt.scatter, "PetalLengthCm", "PetalWidthCm").add_legend()
plt.show()
```

* Box plot for Petal Length

```python
sns.boxplot(x="Species", y="PetalLengthCm", data=iris_df)
plt.show()
```

* Box plot for Petal Width

```python
sns.boxplot(x="Species", y="PetalWidthCm", data=iris_df)
plt.show()
```

* Box plot for Sepal Length

```python
sns.boxplot(x="Species", y="SepalLengthCm", data=iris_df)
plt.show()
```

* Box plot for Sepal Width

```python
sns.boxplot(x="Species", y="SepalWidthCm", data=iris_df)
plt.show()
```

* Facet grid for Petal Length

```python
sns.FacetGrid(iris_df, hue="Species", size=6) \
   .map(sns.kdeplot, "PetalLengthCm") \
   .add_legend()
plt.show()
```

* Facet grid for Petal Width

```python
sns.FacetGrid(iris_df, hue="Species", size=6) \
   .map(sns.kdeplot, "PetalWidthCm") \
   .add_legend()
plt.show()
```

* Facet grid for Sepal Length

```python
sns.FacetGrid(iris_df, hue="Species", size=6) \
   .map(sns.kdeplot, "SepalLengthCm") \
   .add_legend()
plt.show()
```

* Facet grid for Sepal Width

```python
sns.FacetGrid(iris_df, hue="Species", size=6) \
   .map(sns.kdeplot, "SepalWidthCm") \
   .add_legend()
plt.show()
```

* Pair plot for the Iris data

```python
sns.pairplot(iris_df.drop("Id", axis=1), hue="Species", size=3)
plt.show()
```
